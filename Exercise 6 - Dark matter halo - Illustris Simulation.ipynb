**Note:** To simplify things, the following function was provided by the illustris project.

It takes a path (such as 'url' above) and returns either the header information, or downloads binary data and returns the file name.

(from the illustris documentation [here](http://www.illustris-project.org/data/docs/api/) )

```python
def get(path, params=None):
    # make HTTP GET request to path
    headers = {"api-key":"8a8338834d046c14e6d8c15c24f70dda"}
    r = requests.get(path, params=params, headers=headers)

    # raise exception if response code is not HTTP SUCCESS (200)
    r.raise_for_status()

    if r.headers['content-type'] == 'application/json':
        return r.json() # parse json responses automatically

    if 'content-disposition' in r.headers:
        filename = r.headers['content-disposition'].split("filename=")[1]
        with open(filename, 'wb') as f:
            f.write(r.content)
        return filename # return the filename string

    return r
```

Determine the 
   density profiles $\rho = \frac{d N}{dV}$ for the halo at these three redshifts.
   $dN$ is the number of objects in the Volume $dV$.
   Remember: *plt.hist* can be used to build histograms.

   Can you find a redshift dependence?

```python
id = 75
nbins = 25
rmin = 0.
rmax = 100. 
xbins = np.zeros((3, nbins)) #three differnt redshifts
hist = 0.*xbins #initalize volume density
ehist = 0.*xbins # uncertainty of volume density
vol = 0.*xbins # define the volume in each bin

%matplotlib inline

# Loop over snapshots
for i in range(len(snapshot)):
    sub, saved_filname = getsnapshot(id, snapshot[i]) # read in snapshot
    with h5py.File(saved_filename) as f: # read coordinates
        dx = f['PartType1']['Coordinates'][:,0] - sub['pos_x']
        dy = f['PartType1']['Coordinates'][:,1] - sub['pos_y']
        dz = f['PartType1']['Coordinates'][:,2] - sub['pos_z']
    rr = np.sqrt(dx**2. + dy**2. +dz **2.) # get radii 
    # count number in every bin
    # x binedges
    count, x, ax = plt.hist(rr, bins=nbins, range=(0, rmax), alpha=0.5) # create histogram
    # obtain bin centers mean of upper edge and lower edge
    xbins[i, :] = 0.5*(x[1:] + x[:-1])
    # volume of each bin: outer bin edge sphere - inner bin edge sphere
    vol[i,:] = 4.*np.pi/3.*(x[1:]**3. - x[:-1]**3.)
    # divie by volume of bin volume 
    hist[i,:] = count/vol[i,:]
    # uncertainty poisson uncertainty
    ehist[i,:] = np.sqrt(count)/vol[i,:]
    
# Should give histogram plot here with three different snapshots   
```

```python
for i in range(len(snapshot)):
    plt.errorbar(xbins[i,:],hist[i,:],yerr=ehist[i,:])
plt.xscale('log')
plt.yscale('log')
plt.xlim(2,100)
plt.ylim(1e-6,10)
plt.legend(['z=1.0','z=0.5','z=0.0'])
```

Fit the functional form

    $\rho(r)=\frac{A}{\frac{r}{r_s} (1+\frac{r}{r_s})^n}$

   with n=3 or n=4 to the density profiles. Choose "sensible" fitting intervals.

```python
from scipy.optimize import curve_fit
def nfw3(r,a,rs):
    nfw=a/((r/rs)*(1.+(r/rs))**3.)
    return nfw

def nfw4(r,a,rs):
    nfw=a/((r/rs)*(1.+(r/rs))**4.)
    return nfw

labels = ['z=1','z=0.5','z=0']
fig, axarr = plt.subplots(1,3)
for i in range(len(snapshot)):
    valid = ehist[i,:]>0
    popt3,pcov3 = curve_fit(nfw3,xbins[i,valid],hist[i,valid],sigma=ehist[i,valid])
    popt4,pcov4 = curve_fit(nfw4,xbins[i,valid],hist[i,valid],sigma=ehist[i,valid])
    axarr[i].errorbar(xbins[i,:],hist[i,:],yerr=ehist[i,:])
    axarr[i].set_xscale('log')
    axarr[i].set_yscale('log')
    if i>0:
        a = plt.setp(axarr[i].get_yticklabels(), visible=False)
    axarr[i].plot(xbins[i,valid],nfw3(xbins[i,valid],*popt3))
    axarr[i].plot(xbins[i,valid],nfw4(xbins[i,valid],*popt4))
    axarr[i].legend(('n=3','n=4',labels[i]))
    print(popt3,popt4)
```

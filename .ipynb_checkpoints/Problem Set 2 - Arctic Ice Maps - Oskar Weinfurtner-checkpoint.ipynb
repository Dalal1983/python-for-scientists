## Part 2 - reading in multiple maps (10 points)

We now want to make a plot of the ice concentration over time. Reading in a single map is easy, but since we have 137 maps, we do not want to read them all in individually by hand. **Write a loop over all the available files, and inside the loop, read in the data to a variable (e.g. ``data``), and also extract the year, month, and day as integer values (e.g. ``year``, ``month``, and ``day``)**. Then, also inside the loop, **compute a variable ``time``** which is essentially the fractional time in years (so 1st July 2012 is 2012.5). You can assume for simplicity that each month has 30 days - this will not affect the results later. Finally, also **compute for each file the total number of pixels that have a value above 50%**. After the loop, **make a plot of the number of pixels with a concentration above 50% against time**.

You will likely notice that the ticks are in a strange format, where they are given in years since 2006, but you can change this with the following code:

    >>> from matplotlib.ticker import ScalarFormatter
    >>> plt.gca().xaxis.set_major_formatter(ScalarFormatter(useOffset=False))

```python
# Loop over files and read data

import os
current_directory = os.getcwd()
file_directory = current_directory + "/data/ice_data"
file_directory

# create numpy array for dates and time
dates = np.zeros((136, 3), dtype=np.int) #one row consists of year, month, day 
time = np.zeros(136) # time stores floats
counts = np.zeros(136) # counts stores number of datapoints above 50%

count = 0
for root, dirs, filenames in os.walk(file_directory):
    for f in filenames:
        
        # get the dates out of filenames:
        year = int(f.split('.')[0][0:4])
        month = int(f.split('.')[0][4:6])
        day = int(f.split('.')[0][6:])
        
        # save dates in dates array
        dates[count][0] = year
        dates[count][1] = month
        dates[count][2] = day
 
        # compute fractional time
        time[count] = year + (month+(day/30))/12
        
        # Read in the current file
        data = np.load(os.path.join(root, f))
        counts[count] = np.count_nonzero(data > 50)
        
        count += 1
```

```python
# Create a plot of the number of pixels with a concentration above 50% against time

fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([0.13,0.13,0.8,0.8])
ax.set_title("Plot 2: Number of pixels with a concentration above 50% against time")
ax.plot(time, counts, color='b')
ax.set_xlabel(r"Time ")
ax.set_ylabel(r"Number of Pixels")
#ax.legend(title="Legende", borderpad=1.0, borderaxespad=1, loc='best', shadow='true', fontsize='small')
ax.grid(True)
ax.set_ylim((0, 100000))
from matplotlib.ticker import ScalarFormatter
plt.gca().xaxis.set_major_formatter(ScalarFormatter(useOffset=False))

# Save the plot as pdf file
#plt.savefig('pset2_2.pdf', format='pdf')
```

**Describe what you see in the plot**.


As the title indicates, we can see the number of pixels with a concentration above 50% against time, which is a relatively good measure for how much ice-surface there is over time. The graph shows periodic behavior, in particular there are local maxima in the in the first third of a year and local minima near the end of every year. This indicates that the Arctic ice melts during the summer and reaches it's minimum surface in spetember to october, then the ice surface grows again and reaches a maximum in march to april. The global maximum (in the plot) was in early 2008 and similarly high in early 2009, the maximum of 2011 is a bit smaller. The global minimum is found in late 2007 and similarly low in late 2011. Interesting is the the high value in late 2006, which shows the greatest minimum. Now, loop over the files again as before, but this time, for each file, **compute the total area where the concentration of ice is 99% or above. Make a new plot showing the area of >99% ice concentration against time.**

```python
# Loop over files and read data
# Note: this cell takes a while to run, 1 min 3 sec on my machine :D
# cell is finished when counter reaches 136

import os
current_directory = os.getcwd()
file_directory = current_directory + "/data/ice_data"
file_directory

# Read in the pixel data
pixel = np.load('data/ice_data_area.npy')

# create numpy array for dates and time
dates = np.zeros((136, 3), dtype=np.int) #one row consists of year, month, day 
time = np.zeros(136) # time stores floats
area = np.zeros(136) # counts stores number of datapoints above 99%

count = 0
for root, dirs, filenames in os.walk(file_directory):
    for f in filenames:
        print(count) # count can be commented out, but makes sure that the computer is still alive
                     # the print of count serves as prgress-bar, it has to reach 136 for the computation to finish
        
        # get the dates out of filenames:
        year = int(f.split('.')[0][0:4])
        month = int(f.split('.')[0][4:6])
        day = int(f.split('.')[0][6:])
        
        # save dates in dates array
        dates[count][0] = year
        dates[count][1] = month
        dates[count][2] = day
 
        # compute fractional time
        time[count] = year + (month+(day/30))/12
        
        # Read in the current file
        data = np.load(os.path.join(root, f))
        
        # Set NaN-Values to Zero
        data[np.isnan(data)] = 0
        
        # Go through data and sum pixel-area where concentraion is greater 99%, save in area-numpy-array
        for i in range(550):
            for j in range(500):
                if(data[i][j] > 99.0):
                    area[count] += pixel[i][j]
        count += 1
```

```python
# Create a plot of the number of pixels with a concentration above 50% against time

fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([0.13,0.13,0.8,0.8])
ax.set_title("Plot 4: Area of ice with a concentration above 99% against time")
np.arange(136)
ax.plot(time, area, color='b')
ax.set_xlabel(r"Time ")
ax.set_ylabel(r"Area of ice in $km^2$")
#ax.legend(title="Legende", borderpad=1.0, borderaxespad=1, loc='best', shadow='true', fontsize='small')
ax.grid(True)
#ax.set_ylim((0, 20000))
from matplotlib.ticker import ScalarFormatter
plt.gca().xaxis.set_major_formatter(ScalarFormatter(useOffset=False))
plt.gca().yaxis.set_major_formatter(ScalarFormatter(useOffset=False))

# Save the plot as pdf file
#plt.savefig('pset2_4.pdf', format='pdf')
```

## Part 4 - yearly averages (4 points)

**Compute average ice concentration maps for 2006 and 2011, and plot them side by side.**

In total, there are 23 maps for 2006 and 17 maps for 2011. We'll compute the averages by summing the values for individual pixels in arrays and then dividing by the number of maps.

```python
# Set up for file handling

import os
current_directory = os.getcwd()
file_directory = current_directory + "/data/ice_data"
file_directory

# create numpy array for dates and time
av06 = np.zeros((550, 500)) #array for values of 2006
av11 = np.zeros((550, 500)) #array for values of 2011

count = 0
count06 = 0.0
count11 = 0.0
for root, dirs, filenames in os.walk(file_directory):
    for f in filenames:
        #print(count)       
        # if map belongs to 2006, read in and add up numbers to array
        if(int(f.split('.')[0][0:4]) == 2006):
            current = np.load(os.path.join(root, f)) #read in
            current[np.isnan(current)] = 0 # 'clean' NaN values
            av06 += current # add numpy matrices
            count06 += 1 # increment counter for averaging
        
        elif(int(f.split('.')[0][0:4]) == 2011):
            current = np.load(os.path.join(root, f)) #read in 
            current[np.isnan(current)] = 0 # 'clean' NaN values
            av11 += current # add numpy matrices
            count11 += 1 # increment counter for averaging
        
        count += 1
        
# reading completed, now we have to divide for the averages:
av06 = av06/count06
av11 = av11/count11
```
